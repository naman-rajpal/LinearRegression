```
import numpy as np
```

```
class LinearRegression :
    def __init__(self,lr=0.0001,epoch=10000):
        self.lr=lr
        self.epoch=epoch
        self.weights=None
        self.base=None
    def fit(self,X,y):
        self.X_train=X
        self.Y_train=y
        number_of_samples,number_of_features=X.shape
        self.weights=np.zeros(number_of_features)
        self.base=0
        for _ in range(self.epoch):
            y_predicted=np.dot(X,self.weights) + self.base

            dw=1/(number_of_samples) * np.dot(X.T,(y_predicted-y))
            db=1/(number_of_samples) * np.sum(y_predicted-y)

            self.weights -= dw*self.lr
            self.base-=db*self.lr
    def predict(self,X):
        y_predicted=np.dot(X,self.weights) + self.base
        return y_predicted
```

```
import pandas as pd

from sklearn.model_selection import train_test_split
raw_data=pd.read_csv('/content/insurance.csv')
train_data=raw_data.drop(['sex','children','smoker','region','charges'],axis=1)
labels=raw_data['charges']
X_train,X_test,Y_train,Y_test=train_test_split(train_data,labels,test_size=0.2,random_state=1234)
```

```
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
predictions=regressor.predict(X_test)
acc=1/len(Y_test) * np.sqrt(np.sum(predictions-Y_test)**2)
print(acc)
```
